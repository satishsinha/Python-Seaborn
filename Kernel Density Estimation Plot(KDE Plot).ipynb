```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head()
```

```python
df = data[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income']]
df.head()
```

```python
## Basic KDE Plot for 'Total' column
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', data=df)
```

```python
## Create KDE plot for all numeric values in DataFrame
plt.figure(figsize= (8, 5))
sns.kdeplot(data=df)
```

```python
## Adjust the smooting using bw_adjust
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', data=df, bw_adjust=0.1)
```

```python
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', data=df, bw_adjust=0.3)
```

```python
## Group KDE on a category variable and also use 'multiple' attribute
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', hue='Payment', data=df, bw_adjust=0.3, multiple='stack')
```

```python
## Use log scaling to map the variable in KDE
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', hue='Payment', data=df, log_scale=True, multiple='stack')
```

```python
## Change the styling of KDE using linewidth, palette, alpha etc
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', hue='Payment', data=df, log_scale=True, multiple='stack', linewidth=5, palette='Accent_r', alpha=0.1)
```

```python
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Total', hue='Payment', data=df, log_scale=True, multiple='stack', linewidth=5, palette='Accent_r', alpha=0.8)
```

```python
## Create Bivariate KDE
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Unit price', y='gross income', data=df)
```

```python
plt.figure(figsize= (10, 5))
sns.kdeplot(x='Unit price', y='gross income', hue='Gender', data=df, fill=True, thresh=0.2)
```
