```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
df = data[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income', 'Tax 5%']]
df.head(3)
```

```python
## Basic Rug Plot
plt.figure(figsize =(15, 5))
sns.rugplot(data=df, x='Unit price')
```

```python
## Create RUG Plot with two variables
plt.figure(figsize =(15, 5))
sns.rugplot(data=df, x='Unit price', y='gross income')
```

```python
## Group it by categorical variable using 'hue'
plt.figure(figsize =(15, 5))
sns.rugplot(data=df, x='Unit price', y='gross income', hue='Gender')
```

```python
## Change the height of Rugs in RUG PLOT
plt.figure(figsize =(15, 5))
sns.rugplot(data=df, x='Unit price', y='gross income', hue='Gender', height=0.2)
```

```python
plt.figure(figsize =(10, 8))
sns.rugplot(data=df, x='Unit price', y='gross income', hue='Gender', height=0.1)
sns.kdeplot(data=df, x='Unit price', y='gross income', hue='Gender', fill= True, thresh=0.5)
```
