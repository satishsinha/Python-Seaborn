```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('mart_linePlot.xlsx')
```

```python
data.head(3)
```

```python
## Basic reg plot
plt.figure(figsize= (15, 5))
sns.regplot(data=data, x='Outlet_Year', y='Sales')
```

```python
plt.figure(figsize= (15, 5))
sns.regplot(data=data, x='Outlet_Year', y='Sales', line_kws=dict(color='red', linestyle= '--'), 
            scatter_kws=dict(marker='o', color='green', s=100))
```
