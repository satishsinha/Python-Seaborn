```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(10)
```

```python
## Horizontal Bar Plot
plt.figure(figsize =(10, 5))
sns.countplot(y="City", data=data)
```

```python
plt.figure(figsize =(10, 5))
sns.countplot(y="Product line", data=data)
```

```python
## Add hue to get count of two categories i.e Gender and Product Line
plt.figure(figsize =(18, 5))
sns.countplot(x='Product line', hue='Gender', data=data)
```

```python
## Using different color palette
plt.figure(figsize =(18, 5))
sns.countplot(x='Product line', hue='Gender', data=data, palette='brg')
```

```python
## Change style with facecolor, linewidth and edgecolor
plt.figure(figsize =(10, 5))
plt.xlabel('Count', fontweight='bold')
plt.ylabel('Product line', fontweight='bold')
sns.countplot(y="Product line", data=data, facecolor=(1,0,1,0), linewidth= 1, edgecolor= sns.color_palette('dark', 3))
```

```python
## Basic Bar Plot
data.head(2)
```

```python
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', data=data)
```

```python
## Add hue in Bar Plot
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, palette='pastel')
```

```python
## Horizontal BAR Plot
plt.figure(figsize= (10,8))
sns.barplot(x='Total', y='Product line',  hue='Gender', data=data, palette='dark')
```

```python
data['Product line'].sort_values(ascending=True).unique()
```

```python
## Plot the BAR in a given order - in this order this Product line alphabetical order and also hue_order

plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, palette='dark', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle', 'Sports_and_travel'], hue_order=['Male', 'Female'])
```

```python
## Add CAP in error bar
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, palette='bright', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle', 'Sports_and_travel'], hue_order=['Male', 'Female'], capsize=0.5)
```

```python
## Remove Error BAR where ci stands for Class Interval(The `ci` parameter is deprecated. Use `errorbar=None` for the same effect)
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, palette='colorblind', hue_order=['Male', 'Female'], errorbar=None)
```

```python
## Change bar color using color attributes(Setting a gradient palette using color= is deprecated and will be removed in v0.14.0. 
## Set `palette='dark:red'` for the same effect)
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, hue_order=['Male', 'Female'], palette='dark:red')
```

```python
## Use of saturation attributes - to increase value of saturation color saturation will increase
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, hue_order=['Male', 'Female'], palette='pastel', saturation=0.3)
```

```python
## Change default Aggregation method using Estimator parameter - y axis value has changed you can also use for example np.median etc
plt.figure(figsize= (18,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=data, hue_order=['Male', 'Female'], 
            palette='YlGnBu', saturation=0.3, estimator=sum)
```
