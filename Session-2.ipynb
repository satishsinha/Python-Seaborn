```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(10)
```

```python
## Horizontal Bar Plot
plt.figure(figsize =(10, 5))
sns.countplot(y="City", data=data)
```

```python
plt.figure(figsize =(10, 5))
sns.countplot(y="Product line", data=data)
```

```python
## Add hue to get count of two categories i.e Gender and Product Line
plt.figure(figsize =(18, 5))
sns.countplot(x='Product line', hue='Gender', data=data)
```

```python
## Using different color palette
plt.figure(figsize =(18, 5))
sns.countplot(x='Product line', hue='Gender', data=data, palette='brg')
```

```python
## Change style with facecolor, linewidth and edgecolor
plt.figure(figsize =(10, 5))
plt.xlabel('Count', fontweight='bold')
plt.ylabel('Product line', fontweight='bold')
sns.countplot(y="Product line", data=data, facecolor=(1,0,1,0), linewidth= 1, edgecolor= sns.color_palette('dark', 3))
```
