```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('mart_linePlot.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
data.shape
```

```python
df = data.pivot_table(index='Outlet_Year', columns='Outlet_Size', values='Sales')
df
```

```python
##Basic Heat Map
plt.figure(figsize= (10, 8))
sns.set_style('white')
sns.heatmap(df)
```

```python
plt.figure(figsize= (10, 8))
sns.set_style('white')
sns.heatmap(df, annot=True)
```

```python
plt.figure(figsize= (10, 8))
sns.set_style('white')
sns.heatmap(df, annot=True, fmt='.0f')
```
