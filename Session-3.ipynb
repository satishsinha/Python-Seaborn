```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(2)
```

```python
## BOX PLOT - provide a summary of supplied data, which includes information like: mean,median,minimum,maximum,1st Quartile,3rd Quartile,Outliers
```

```python
## Create a basic BOX Plot for one numeric variable
sns.boxplot(x='Total', data=data, width =0.9)
```

```python
sns.boxplot(y='Total', data=data, width =0.2)
```

```python
## Create a basic BOX Plot on one numeric variable by a categorical variable
sns.boxplot(x='Payment', y='Total', data=data)
```

```python
## Create a basic BOX Plot on one numeric variable by Two categorical variable using 'hue' attributes
plt.figure(figsize=(10,5))
sns.boxplot(x='Payment', y='Total', hue='Gender', data=data)
```

```python
## Add mean marker in the BOX PLOT using 'Showmeans' attributes and change its style using meanprops
plt.figure(figsize=(10,5))
sns.boxplot(x='Payment', y='Total', hue='Gender', data=data, palette='dark', showmeans= True)
```

```python
plt.figure(figsize=(15,6))
plt.title("Gender & Payment Type wise analysis", fontweight="bold")
plt.xlabel("Payment Type", fontweight="bold")
plt.ylabel("Total Payment", fontweight="bold")
sns.boxplot(x='Payment', y='Total', hue='Gender', data=data, palette='pastel', showmeans= True, 
            meanprops={"marker":"o", "markerfacecolor":"white", "markersize": 10, "markeredgecolor":"red" })
```

```python
## Make BOX PLOT horizontal
plt.figure(figsize=(6,8))
plt.title("Gender & Payment Type wise analysis", fontweight="bold")
plt.xlabel("Payment Type", fontweight="bold")
plt.ylabel("Total Payment", fontweight="bold")
sns.boxplot(x='Total', y='Payment',  hue='Gender', data=data, palette='pastel', showmeans= True, 
            meanprops={"marker":"o", "markerfacecolor":"white", "markersize": 10, "markeredgecolor":"red" })
```

```python
## Change Palette, line width etc
plt.figure(figsize=(15,6))
plt.title("Gender & Payment Type wise analysis", fontweight="bold")
plt.xlabel("Payment Type", fontweight="bold")
plt.ylabel("Total Payment", fontweight="bold")
sns.boxplot(x='Payment', y='Total', hue='Gender', data=data, palette='BuPu', linewidth= 4, showmeans= True, 
            meanprops={"marker":"o", "markerfacecolor":"white", "markersize": 6, "markeredgecolor":"red" })
```

```python
## Create BOX PLOT for each of the Numeric Variable in the dataframe
plt.figure(figsize=(15,5))
sns.boxplot(data=data)
```
