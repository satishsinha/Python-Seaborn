```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('mart_train_sample.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
## Basic Scatter Plot
plt.figure(figsize = (10,6))
sns.scatterplot(data=data, x='Item_MRP', y='Sales')
```

```python
data.head(3)
```

```python
## Grouping basis on a categorical variable by 'hue' 
plt.figure(figsize = (10,6))
plt.grid(True, linestyle='--', color='green', linewidth=0.7)
sns.scatterplot(data=data, x='Item_MRP', y='Sales', hue='Outlet_Size')
```

```python
## Grouping basis on a categorical variable using STYLE and styling markers as well
plt.figure(figsize = (10,6))
plt.grid(True, linestyle='--', color='green', linewidth=0.7)
sns.scatterplot(data=data, x='Item_MRP', y='Sales', style='Outlet_Size', markers={"Small":"*", "Medium": "^", "High": "o"})
```

```python
data.head()
```

```python
## Grouping basis on a categorical variable using HUE & STYLE both together
plt.figure(figsize = (10,6))
plt.grid(True, linestyle='--', color='green', linewidth=0.7)
sns.scatterplot(data=data, x='Item_MRP', y='Sales', hue='Outlet_Size', style='Outlet_Location_Type', markers={"Tier 3":"*", "Tier 2": "^", "Tier 1": "o"})
```

```python
## Change the Marker size using S agrument
plt.figure(figsize = (10,6))
plt.grid(True, linestyle='--', color='green', linewidth=0.7)
sns.scatterplot(data=data, x='Item_MRP', y='Sales', hue='Outlet_Size', style='Outlet_Location_Type', 
                markers={"Tier 3":"*", "Tier 2": "^", "Tier 1": "o"}, s=300, edgecolor='black')
```
