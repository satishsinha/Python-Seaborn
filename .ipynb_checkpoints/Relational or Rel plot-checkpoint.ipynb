```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('mart_linePlot.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
## Basic rel plot
plt.figure(figsize= (15, 5))
sns.relplot(data=data, x='Outlet_Year', y='Sales', kind='line', hue='Outlet_Size', errorbar=None, col='Item_Type', col_wrap=4)
```

```python
plt.figure(figsize= (15, 10))
sns.relplot(data=data, x='Outlet_Year', y='Sales', s=100, hue='Outlet_Size', col='Tier', palette=['green', 'red', 'yellow'])
```
