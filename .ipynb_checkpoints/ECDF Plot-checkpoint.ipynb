```python
## Empirical Cumulative Distribution Functions
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
## I want to create Hitogram on the basis of particulars column, so i create an another dataframe
df = data[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income', 'Rating']]
df.head()
```

```python
## Basic ECDF plot
plt.figure(figsize= (10, 6))
plt.grid(True)
sns.ecdfplot(x='gross income', data=df)
```

```python
plt.figure(figsize= (10, 6))
plt.grid(True)
sns.ecdfplot(x='gross income', hue='Gender', data=df)
```

```python
plt.figure(figsize= (10, 6))
plt.grid(True)
sns.ecdfplot(x='gross income', hue='Gender', stat='count', data=df)
```

```python
## With Y Axis
plt.figure(figsize= (10, 6))
plt.grid(True)
sns.ecdfplot(y='gross income', hue='Gender', stat='count', data=df)
```
