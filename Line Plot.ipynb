```python
## Show the relation between two variables, majorly used for time series analysis. How the value of varibale changes over the time
```

```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('mart_linePlot.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(3)
```

```python
## Basic Line Plot and try different estimator
plt.figure(figsize = (10, 5))
sns.lineplot(data=data)
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales')
```

```python
## The `ci` parameter is deprecated. Use `errorbar=None` for the same effect.
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales', errorbar=None)
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales', estimator='sum')
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales', estimator=None)
```

```python
## Grouping using 'hue' and using palette
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales', hue='Outlet_Size', errorbar=None, palette=['red', 'green', 'blue'])
```

```python
## Grouping using Style, use different marker and Dashes
plt.figure(figsize = (10, 5))
sns.lineplot(data=data, x='Outlet_Year', y='Sales', hue='Outlet_Size', style='Outlet_Size', errorbar=None, palette=['red', 'green', 'blue'], marker=True)
```
