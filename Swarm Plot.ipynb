```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
## Create basic SWARM Plot on all columns
# plt.figure(figsize=(8, 5))
# sns.swarmplot(data=data)
```

```python
## Create basic SWARM Plot on 'Total' columns
plt.figure(figsize=(8, 5))
sns.swarmplot(y='Total', data=data)
```

```python
## Swarm plot group by CATEGORIES
plt.figure(figsize=(8, 5))
sns.swarmplot(x='Payment', y='Total', data=data, color='green')
```

```python
## Show 'hue' separatly on the categorical axes
plt.figure(figsize=(8, 5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', dodge=True, data=data)
```

```python
## Styling SWARM Plot using marker, size, color, edgecolor etc
plt.figure(figsize=(14, 5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', dodge=True, data=data, marker='*', size=7, palette='dark:green')
```

```python
## Overlay Swarm plot on a Box Plot 
plt.figure(figsize=(15, 5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', dodge=True, data=data, marker='^', size=7, palette='dark:green')
sns.boxplot(x='Payment', y='Total', hue='Gender',data=data, palette='pastel')
```

```python
## Overlay Swarm plot on a Violin Plot 
plt.figure(figsize=(15, 5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', dodge=True, data=data, marker='^', size=7, palette='dark:white')
sns.violinplot(x='Payment', y='Total', hue='Gender',data=data, palette='PuBu')
```
