```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(2)
```

```python
## Make a basic Violin Plot
plt.figure(figsize =(10,5))
sns.violinplot(y='Total', data=data)
```

```python
## Create a violin plot using two categorical and One numerical variable and use split
plt.figure(figsize =(10,5))
sns.violinplot(x='Gender', y='City', data=data)
```

```python
plt.figure(figsize =(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data)
```

```python
plt.figure(figsize =(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data, split=True)
```

```python
## Chnage the box for violin plot into Horizontal
plt.figure(figsize =(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data, split=True, inner='quartile')
```

```python
data.head(2)
```

```python
## Draw line for each observation in VIOLIN PLOT
plt.figure(figsize=(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data, split=True, inner='stick' )
```

```python
## Change the amount of smoothing using 'bw' attributes
## The `bw` parameter is deprecated in favor of `bw_method`/`bw_adjust`.Setting `bw_method=0.2`, but please see docs for the new parameters
## and update your code. This will become an error in seaborn v0.15.0.

plt.figure(figsize=(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data, split=True, inner='quartile', bw_method=0.2 )
```

```python
## Cut out the extreme values
plt.figure(figsize=(10,5))
sns.violinplot(x='Payment', y='Total', hue='Gender', data=data, split=True, inner='quartile', bw_method=0.2, cut=0 )
```
