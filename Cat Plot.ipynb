```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(2)
```

```python
## Basic CAT PLOT
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', data=data, kind='box')
```

```python
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', data=data, kind='bar')
```

```python
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', data=data, kind='violin')
```

```python
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', hue='Gender', data=data, kind='swarm')
```

```python
## replace 'hue' to 'col' - and check the changes
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', col='Gender', data=data, kind='bar')
```

```python
data.head()
```

```python
##Adding 'row' attributes
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', col='Gender', data=data, kind='bar', row='Customer type')
```

```python
## Adding color, palette etc
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', col='Gender', data=data, kind='bar', row='Customer type', palette='CMRmap_r')
```

```python
plt.figure(figsize=(8,5))
sns.catplot(x='Payment', y='Total', col='Gender', data=data, kind='box', row='Customer type', color='green')
```
