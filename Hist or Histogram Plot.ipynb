```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head()
```

```python
## I want to create Hitogram on the bais of particulars column, so i create an another dataframe
df = data[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income']]
df.head()
```

```python
## Create a basic Hist Plot on X axis
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='gross income')
```

```python
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total')
```

```python
## Create a basic Hist Plot on Y axis
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, y='Total')
```

```python
## Change the Bins width using 'Bandwidth' argument
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='gross income', binwidth=0.9)
```

```python
## Change the number of bins and interval
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', bins=np.arange(0, 1100, 50))
```

```python
## To check how every bin size contributes in Histogram Plot
plt.figure(figsize=(15,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', bins=np.arange(0, 1100, 50))
plt.xticks(np.arange(0, 1100, 50))
```

```python
## Combine with KDE
plt.figure(figsize=(10,5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', kde=True)
```

```python
## Use a categorical variable in 'HUE' and 'STACK' it uses MULTIPLE agrument
plt.figure(figsize=(10,5))
## plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', hue='Payment')
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='gross income', hue='Gender', bins=np.arange(0, 60, 5), kde=True)
```

```python
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', hue='Gender', binwidth=50, multiple='stack')
```

```python
## Make a step/poly plot using 'ELEMENT' attribute and change the fill
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', element='step')
```

```python
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', element='step', fill=False)
```

```python
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', element='poly')
```

```python
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', hue='Gender', element='poly', fill=True)
```

```python
## Use Categorical varible and shrink it
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Payment', shrink=0.5)
```

```python
## count, density, probability, frequency
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', stat='probability')
```

```python
## Create a Bivariate Histogram
plt.figure(figsize=(15, 5))
plt.grid(True, linestyle='--', color='gray', linewidth=0.7)
sns.histplot(data=df, x='Total', y='gross income')
```
