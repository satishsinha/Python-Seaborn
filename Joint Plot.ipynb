```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
## Import data frame
data = pd.read_excel('supermarket_sales.xlsx')
```

```python
## Complete information of Data Frame
data.info()
```

```python
data.head(2)
```

```python
df = data[['Gender', 'Unit price', 'Quantity', 'Tax 5%', 'Total', 'Payment', 'gross income', 'Rating']]
df
```

```python
## I want to create Hitogram on the basis of particulars column, so i create an another dataframe
df = data[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income', 'Rating']]
df.head()
```

```python
## Basic Joint Plot
plt.figure(figsize = (10,5))
sns.jointplot(data=df, x='gross income', y='Unit price')
```

```python
## Change the kind of 'scatter' or 'kde' or 'hist' or 'hex' or 'reg' or 'resid'
plt.figure(figsize = (10,5))
sns.jointplot(data=df, x='gross income', y='Unit price', kind='kde')
```

```python
df.head(2)
```

```python
## Grouping based on categorical variable
plt.figure(figsize = (10,5))
sns.jointplot(data=df, x='gross income', y='Unit price', hue='Gender', kind='scatter')
```

```python
## Formatting All and Individual Plot
plt.figure(figsize = (10,5))
sns.jointplot(data=df, x='gross income', y='Unit price', joint_kws= dict(marker='*', color= 'red'), marginal_kws=dict(color='green'))
```
